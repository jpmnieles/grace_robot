```python
left_img = cv2.imread(os.path.join(os.getcwd(), 'dataset', '231107_subject0','20231107_142441_385159_left_eye.png'))
gray_left_img = cv2.cvtColor(left_img, cv2.COLOR_BGR2GRAY)
right_img = np.array(cv2.imread(os.path.join(os.getcwd(), 'dataset', '231107_subject0','20231107_142441_385159_right_eye.png')))
gray_right_img = cv2.cvtColor(right_img, cv2.COLOR_BGR2GRAY)
```

```python
def mean_squared_error(y_true, y_pred):
    """
    Calculate the Mean Squared Error (MSE) between the true and predicted values.

    Arguments:
    y_true -- numpy array, true values
    y_pred -- numpy array, predicted values

    Returns:
    mse -- float, mean squared error
    """
    # Check if the inputs have the same shape
    assert y_true.shape == y_pred.shape, "Shapes of y_true and y_pred do not match."

    # Calculate the squared difference
    squared_diff = np.square(y_true - y_pred)

    # Calculate the mean
    mse = np.mean(squared_diff)

    return mse
```

```python
mean_squared_error(gray_left_img, gray_right_img)
```
